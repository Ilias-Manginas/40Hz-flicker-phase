```python
# Imports

import matplotlib
import helpers.helper_functions as hf
import mne
import matplotlib.pyplot as plt
import os.path as op
from pathlib import Path

ss = hf.settings_dict()
```

```python
# Epoch parameters
tmin = -0.5
tmax = 4.5
```

```python

for subject_index in ss['subject_idx_list']:

    subject = ss['subject_id_list'][subject_index]
    print("loading raw dataset for subject: ", subject)

    raw_filtered = hf.load_raw_filtered(subject=subject, preload=True)

    # events = mne.find_events(raw_filtered)
    events_fname = op.join(ss['events_dir'], subject + '-eve.fif')
    event_id_list =  ss['event_id_list']

    #events = mne.find_events(raw_filtered)
    events = mne.read_events(events_fname)

    epochs = mne.Epochs(raw_filtered, events, event_id_list, tmin=tmin, tmax=tmax, preload=True)

    save_dir = Path(ss['epochs_dir'])
    save_dir.mkdir(parents=True, exist_ok=True)

    epochs.save(save_dir / f"{subject}-epo.fif", overwrite=True)

    del epochs
    del raw_filtered
```
