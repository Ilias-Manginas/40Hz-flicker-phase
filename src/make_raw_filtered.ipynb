Filter below 38 Hz and above 42 Hz (Since we only care about brain activity at 40 Hz)
Filter for powerline artifacts

```python
# imports
import gc
import matplotlib
import helpers.helper_functions as hf
import mne
import matplotlib.pyplot as plt
import os.path as op
from pathlib import Path

ss = hf.settings_dict()
```

Filtering parameters

```python
l_freq = 38.0
h_freq = 42.0
h_trans_bandwidth = 1.0
l_trans_bandwidth = 1.0


save_dir = Path(ss['raw_filtered_dir'])
save_dir.mkdir(parents=True, exist_ok=True)
```

First load the data and look for bad channels

```python
# load raw datasets one by one
for subject_index in ss['subject_idx_list']:

    subject = ss['subject_id_list'][subject_index]
    print("loading raw dataset for subject: ", subject)

    raw = hf.load_raw(subject=subject_index, preload=True)

    # drop projectors
    raw.info["projs"]
    raw.del_proj()

    raw.pick(picks=['grad', 'stim', 'misc','eeg'])

    psd_before = raw.compute_psd(fmax=130).plot(average=True, amplitude=False, picks="data", exclude="bads", show=False)
    psd_before.suptitle(f"{subject} - Before filtering")

    raw.filter(l_freq=l_freq, h_freq=h_freq, h_trans_bandwidth=h_trans_bandwidth, l_trans_bandwidth=l_trans_bandwidth)

    psd_after = raw.compute_psd(fmax=130).plot(average=True, amplitude=False, picks="data", exclude="bads", show=False)
    psd_after.suptitle(f"{subject} - After filtering")

    plt.show()

    raw.save(save_dir / f"{subject}-raw.fif", overwrite=True)
    del raw

gc.collect()
```
