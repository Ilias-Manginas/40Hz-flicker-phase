```python
# Imports
import math
import helpers.helper_functions as hf
import mne
import os.path as op
from mne.channels import combine_channels
import pandas as pd
from mne.beamformer import make_lcmv, apply_lcmv_epochs
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import hilbert
import helpers.test_circ_plot as circ_plot
import gc
import helpers.stc_helper as stc_helper
import time

ss = hf.settings_dict()
```

```python
initial_time = 2.0
event_id =4
repetitions = 6
time_interval = 0.01
```

```python

for subject_index in ss['subject_idx_list']:
    # loop over each event type
    event_name = str(event_id)
    duty_cycle = ss['event_name_list'][event_id-1]
    subjects_dir = ss['fs_subjects_dir']
    subject = ss['subject_id_list'][subject_index]
    print("loading dataset for subject: ", subject)

    # load hilbert stc data
    phase_diff_voxel_stc_file = Path(ss['hilbert_ref_dir']) / subject / event_name / f"{subject}-event-{event_name}-hilbert-vox-ref-vol.stc"
    phase_diff_voxel_stc = mne.read_source_estimate(phase_diff_voxel_stc_file)

    phase_diff_retina_stc_file = Path(ss['hilbert_ref_dir']) / subject / event_name / f"{subject}-event-{event_name}-hilbert-ret-ref-vol.stc"
    phase_diff_retina_stc = mne.read_source_estimate(phase_diff_retina_stc_file)


    # load forward model
    fwd_fname = op.join(ss['fwd_dir'], subject + '-fwd.fif')
    fwd = mne.read_forward_solution(fwd_fname)
    src = fwd['src']

    # get reference voxel coordinates
    # load hilbert references
    reference_file = Path(ss['hilbert_dir']) / subject / event_name / f"{subject}-event-{event_name}-reference.csv"

    # Read the CSV file
    df = pd.read_csv(reference_file)
    # get voxel number
    voxel = int(df.axes[1][2].split("_")[0])

    coords = stc_helper.extract_voxel_coords_from_stc(stc=phase_diff_retina_stc, voxel=voxel, src=src)


    save_dir = Path(ss['hilbert_dir']) / "reference_video" / subject / event_name
    save_dir.mkdir(parents=True, exist_ok=True)

    for i in range(repetitions):
        circ_plot.plot_circ_stat_map(phase_diff_voxel_stc,
                                                 subject=subject,
                                                 subjects_dir=subjects_dir,
                                                 cut_coords=coords,
                                                 time=initial_time + i*time_interval,
                                                 draw_cross=True,
                                                 title=f"{subject}  |  {duty_cycle} % |  {math.ceil((initial_time + i*time_interval)*1000)} ms  |  voxel as reference",)

        circ_plot.plot_circ_stat_map(phase_diff_retina_stc,
                                                 subject=subject,
                                                 subjects_dir=subjects_dir,
                                                 cut_coords=coords,
                                                 time=initial_time + i*time_interval,
                                                 draw_cross=True,
                                                 title=f"{subject}  |  {duty_cycle} %  |  {math.ceil((initial_time + i*time_interval)*1000)} ms  |  retina as reference",)


    del phase_diff_voxel_stc
    del phase_diff_retina_stc
    gc.collect()
    time.sleep(10)
```
