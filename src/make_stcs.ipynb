```python
# Imports
import gc
import helpers.helper_functions as hf
import mne
import matplotlib.pyplot as plt
import os.path as op
from mne.channels import combine_channels
import pandas as pd
from mne.beamformer import make_lcmv, apply_lcmv_epochs
from pathlib import Path



ss = hf.settings_dict()
```

```python
# Parameters

#data covariance times
# tmax = 3.7 - 1/ss['fs']
# tmin = 0.7
# base_tmin = -0.5 + (1/ss['fs'])
# base_tmax = 0.0

tmax = 3.7
tmin = 0.7
base_tmin = -0.5
base_tmax = 0.0
method = 'empirical' # use shrunk ?

#beamformer parameters
reg=0.05
pick_ori='max-power'
weight_norm='nai'
```

```python


for subject_index in ss['subject_idx_list']:

    subjects_dir = ss['fs_subjects_dir']
    subject = ss['subject_id_list'][subject_index]
    print("loading dataset for subject: ", subject)

    epochs = mne.read_epochs(op.join(ss['epochs_dir'],subject+"-epo.fif"), preload=True)

    # now manking the stc
    epochs.pick('grad')

    #Load forward model
    #TODO: check that the forward models are correct
    fwd = hf.load_fwd_model(subject_index)

    #Source localization

    #make data covariance matrix
    stim_cov = mne.compute_covariance(epochs, tmin=tmin, tmax=tmax, method=method)
    # base_cov = mne.compute_covariance(epochs, tmin=base_tmin, tmax=base_tmax, method=method)
    base_cov = None

    data_cov = stim_cov

    cov_plot = data_cov.plot(info = epochs.info, show= False)
    cov_plot[0].suptitle(f"subject {subject} ", fontsize=16)
    plt.show()

    for event_id in ss['event_id_list']:
        event_name = str(event_id)

        save_dir = Path(ss['stc_dir']) / subject / event_name
        save_dir.mkdir(parents=True, exist_ok=True)

        event_epochs = epochs[event_name]

        sp_filter = make_lcmv(
            event_epochs.info,
            fwd,
            data_cov=data_cov,
            noise_cov=base_cov,
            reg=reg,
            pick_ori=pick_ori,
            weight_norm=weight_norm,
            rank=None,
        )

        stcs = apply_lcmv_epochs(event_epochs, sp_filter)

        for i in range(ss['n_trials']):
            stcs[i].save(save_dir / f"{subject}-{i}-vol.stc" , overwrite=True)

    del stcs
    del epochs

gc.collect()
```
