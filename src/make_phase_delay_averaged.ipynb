```python
# Imports

import helpers.helper_functions as hf
import mne
import os.path as op
from mne.channels import combine_channels
import pandas as pd
from mne.beamformer import make_lcmv, apply_lcmv_epochs
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import hilbert
import helpers.test_circ_plot as circ_plot
import gc
import helpers.stc_helper as stc_helper
import time
import pycircstat2.descriptive as circ_stats
ss = hf.settings_dict()
```

```python
for subject_index in ss['subject_idx_list']:
    for event_id in ss['event_id_list']:
        event_name = str(event_id)
        subject = ss['subject_id_list'][subject_index]
        print("loading dataset for subject: ", subject)

        save_dir = Path(ss['hilbert_ref_dir']) / subject / event_name
        save_dir.mkdir(parents=True, exist_ok=True)

        # --- Load all trials ---
        n_trials = ss['n_trials']
        stc_data = []
        for i in range(n_trials):
            hilbert_stc_file = (
                Path(ss['hilbert_ref_dir']) / subject / event_name
                / f"{subject}-trial{i}-ret-ref-wrp-vol.stc"
            )
            stc = mne.read_source_estimate(hilbert_stc_file)
            stc_data.append(stc.data)

        # stc_data_stacked shape: (n_trials, n_voxels, n_times)
        stc_data_stacked = np.stack(stc_data)

        complex_sum = np.exp(1j * stc_data_stacked).mean(axis=0)
        circ_mean_map_stacked = np.angle(complex_sum)  # [-π, π]

        # --- Save ---
        avg_phase_stc = stc.copy()
        avg_phase_stc.data = circ_mean_map_stacked
        print(f"min  : {circ_mean_map_stacked.min():.4f}")
        print(f"max  : {circ_mean_map_stacked.max():.4f}")
        print(f"mean : {circ_mean_map_stacked.mean():.4f}")
        avg_phase_stc.save(
            save_dir / f"{subject}-event-{event_name}-mean-vol.stc",
            overwrite=True
        )

```
