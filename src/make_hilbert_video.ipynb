```python
# Imports

import helpers.helper_functions as hf
import mne
import os.path as op
from mne.channels import combine_channels
import pandas as pd
from mne.beamformer import make_lcmv, apply_lcmv_epochs
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import hilbert
import helpers.test_circ_plot as circ_plot
import time
import gc
import random
import helpers.stc_helper as stc_helper

ss = hf.settings_dict()
```

```python
#important variables

#1.5 seconds max
tmin = 1.7
tmax = 2.7


event_ids = [4]
subjects = [11]
trials = 4
n_workers = 11
mask = False
```

```python
%matplotlib Agg
for subject_index in subjects:
    # loop over each event type
    for event_id in event_ids:

        event_name = str(event_id)
        duty_cycle = ss['event_name_list'][event_id-1]
        subjects_dir = ss['fs_subjects_dir']
        subject = ss['subject_id_list'][subject_index]
        print("loading dataset for subject: ", subject)

        fwd = hf.load_fwd_model(subject_index)
        src = fwd['src']

        stcs = []
        trial_indexes = []

        for i in random.sample(range(ss['n_trials']), trials):
            hilbert_stc_file = Path(ss['hilbert_dir']) / subject / event_name / f"{subject}-trial-{i}-hilbert-vol.stc"

            stc = mne.read_source_estimate(hilbert_stc_file)

            trial_indexes.append(i)
            stcs.append(stc)



        save_dir = Path(ss['results_dir']) / "video" / "phase"
        save_dir.mkdir(parents=True, exist_ok=True)

        cropped_stcs = []
        for stc in stcs:
            cropped_stc = stc.copy().crop(tmin=tmin, tmax=tmax)
            cropped_stcs.append(cropped_stc)

        circ_plot.plot_multi_circ_stat_map_video(stcs=cropped_stcs,
                                                 src=src,
                                           subjects_dir=subjects_dir,
                                           subject=subject,
                                           fps=5,
                                           dpi=120,
                                           time_indices = None,
                                           output_path=save_dir / f"{subject}-phase-vol-stim.mp4",
                                           title =f"subject: {subject}  |  duty cycle: {duty_cycle} %  |  trials: {trial_indexes}",
                                           nrows=2,
                                           ncols=2,
                                           figsize=(16,9),
                                           n_workers=n_workers,
                                           cut_coords=(0,0,0)
                                           )

        del stcs
        del cropped_stcs
        gc.collect()

        time.sleep(10)
```
